```python
import random

# Define a list of tasks with their durations
tasks = {
    "Task A": 4,
    "Task B": 2,
    "Task C": 3,
    "Task D": 1,
    "Task E": 5,
}

def total_duration(task_order):
    return sum(tasks[task] for task in task_order)

def hill_climbing(initial_order):
    current_order = initial_order
    current_duration = total_duration(current_order)

    while True:
        neighbors = []

        for i in range(len(current_order)):
            for j in range(i + 1, len(current_order)):
                new_order = current_order[:i] + current_order[j:j+1] + current_order[i+1:j] + current_order[i:i+1] + current_order[j+1:]
                neighbors.append(new_order)

        if not neighbors:
            break

        best_neighbor = min(neighbors, key=lambda order: total_duration(order))

        if total_duration(best_neighbor) >= current_duration:
            break

        current_order = best_neighbor
        current_duration = total_duration(current_order)

    return current_order, current_duration

# Initial random order
initial_order = list(tasks.keys())
random.shuffle(initial_order)

best_order, best_duration = hill_climbing(initial_order)

print("Best task order:", best_order)
print("Total duration:", best_duration)
```
