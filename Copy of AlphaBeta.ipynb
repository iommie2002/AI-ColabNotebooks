```python

MAX, MIN = 100, -100  #min & max value limit
def minimax(depth, nodeIndex, maximizingPlayer,values, alpha, beta):

	if depth == 3: #depth for search best solution
		return values[nodeIndex]

	if maximizingPlayer:
		best = MIN
		for i in range(0, 2):
			#search for child nodes value
			val = minimax(depth + 1, nodeIndex * 2 + i,False, values, alpha, beta)
			best = max(best, val)
			alpha = max(alpha, best)

			if beta <= alpha:			# Alpha Beta Pruning
				break

		return best

	else:
		best = MAX
		for i in range(0, 2):
			#search for child nodes value
			val = minimax(depth + 1, nodeIndex * 2 + i,True, values, alpha, beta)
			best = min(best, val)
			beta = min(beta, best)
			if beta <= alpha:   		# Alpha Beta Pruning
				break

		return best

if __name__ == "__main__":

	values = [-30, 80, -10, 90, 55, 21, 36, 85]
	print("Optimal Solution :", minimax(0, 0, True, values, MIN, MAX))
```
