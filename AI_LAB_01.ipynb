```python
#two environments

#A and B in first environment

#C and D in second environment

import random

class Environment(object):
    def __init__(self):
      #initiate the locations and conditions
      #assiging 1 for dirt and 0 for cleanliness

        self.locationCondition = {'A': '0', 'B': '0', 'C': '0', 'D': '0'}

        #randamizing the 4 different locations
        self.locationCondition['A'] = random.randint(0, 1)
        self.locationCondition['B'] = random.randint(0, 1)
        self.locationCondition['C'] = random.randint(0, 1)
        self.locationCondition['D'] = random.randint(0, 1)


#function for environment
class SimpleReflexVacuumAgent(Environment):
    def __init__(self, environment):
        print(environment.locationCondition)
        score = 0  #start from zero

        #there can be a random choice for any of 4 locations
        vacuumLocation = random.choice(['A', 'B', 'C', 'D'])

          #if vaccun cleaner is at location A , then print vaccum cleaner is at A
        if vacuumLocation == 'A':  #condition for location A
            print("Vacuum is randomly placed at Location A")

            #if environment is dirty then print that location
            if environment.locationCondition['A'] == '1':
                print("Location A is Dirty.")

                #clean the  environment and assign the value 0 to location A
                environment.locationCondition['A'] = '0'
                score += 1   #once the  location is cleaned then assign the score as +1 and for dirty location reduce the score by -1
                print("Location A has been Cleaned.")


                if environment.locationCondition['B'] == '1':
                    print("Location B is Dirty.")
                    print("Moving to Location B...")
                    score -= 1
                    environment.locationCondition['B'] = '0'
                    score += 1
                    print("Location B has been Cleaned.")
            else:
                if environment.locationCondition['B'] == '1':
                    print("Location B is Dirty.")
                    score -= 1
                    print("Moving to Location B...")
                    environment.locationCondition['B'] = '0'
                    score += 1
                    print("Location B has been Cleaned.")
        elif vacuumLocation == 'B':   #condition for location B
            print("Vacuum is randomly placed at Location B.")
            if environment.locationCondition['B'] == '1':
                print("Location B is Dirty")
                environment.locationCondition['B'] = '0'
                score += 1
                print("Location B has been Cleaned.")

                if environment.locationCondition['A'] == '1':
                    print("Location A is Dirty")
                    score -= 1
                    print("Moving to Location A")
                    environment.locationCondition['A'] = '0'
                    score += 1
                    print("Location A has been Cleaned.")
            else:
                if environment.locationCondition['A'] == '1':
                    print("Location A is Dirty")
                    print("Moving to Location A")
                    score -= 1
                    environment.locationCondition['A'] = '0'
                    score += 1
                    print("Location A has been Cleaned.")

#room 2 condition starts here

        elif vacuumLocation == 'C': #condition for location C
            print("Vacuum is randomly placed at Location C.")
            if environment.locationCondition['C'] == '1':
                print("Location C is Dirty")
                environment.locationCondition['C'] = '0'
                score += 1
                print("Location C has been Cleaned.")

                if environment.locationCondition['D'] == '1':
                    print("Location D is Dirty")
                    score -= 1
                    print("Moving to Location D")
                    environment.locationCondition['D'] = '0'
                    score += 1
                    print("Location D has been Cleaned.")
            else:
                if environment.locationCondition['D'] == '1':
                    print("Location D is Dirty")
                    print("Moving to Location D")
                    score -= 1
                    environment.locationCondition['D'] = '0'
                    score += 1
                    print("Location D has been Cleaned.")
        elif vacuumLocation == 'D':   #condition for location D
            print("Vacuum is randomly placed at Location D.")
            if environment.locationCondition['D'] == '1':
                print("Location D is Dirty")
                environment.locationCondition['D'] = '0'
                score += 1
                print("Location D has been Cleaned.")

                if environment.locationCondition['C'] == '1':
                    print("Location C is Dirty")
                    score -= 1
                    print("Moving to Location C")
                    environment.locationCondition['C'] = '0'
                    score += 1
                    print("Location C has been Cleaned.")
            else:
                if environment.locationCondition['C'] == '1':
                    print("Location C is Dirty")
                    print("Moving to Location C")
                    score -= 1
                    environment.locationCondition['C'] = '0'
                    score += 1
                    print("Location C has been Cleaned.")

        print(environment.locationCondition)
        print("Performance Measurement: " + str(score))

# Create three different environments
environment1 = Environment()
environment2 = Environment()


# Create SimpleReflexVacuumAgent instances for each environment
vacuum1 = SimpleReflexVacuumAgent(environment1)
vacuum2 = SimpleReflexVacuumAgent(environment2)

#peas : performance: the  ability to increase the  score for  clean location and then reducing by one for dirt
#environemnt : 2 rooms  each with two sections A B C D
#actuators : trace the location and then move to another
#sensors : camera,location tracing
```
