```python
class Graph:
    def __init__(self, adjacency_matrix):
        self.adjacency_matrix = adjacency_matrix

    # Heurestic function returning value
    def heurestic_function(self, node):
        Heurestic_values = {
            'A': 366,
            'Z': 374,
            'S': 253,
            'T': 329,
            'L': 244,
            'O': 380,
            'M': 241,
            'D': 242,
            'R': 193,
            'P': 100,
            'C': 160,
            'F': 176,
            'B': 0
        }

        return Heurestic_values[node]

    # Finding neighbours of current node selected in loop
    def find_all_neighbours(self, node):
        return self.adjacency_matrix[node]

    # A* star algorthm
    def a_star_main(self, start, end):

        # Two arrays for storing explored and not explored nodes
        open_array = set([start])
        close_array = set([])

        # To store distance between nodes
        node_distance = {}
        node_distance[start] = 0

        # To store adjacent nodes name
        ajacent_node_name = {}
        ajacent_node_name[start] = start

        # Run the loop until and unless open_array is empty
        while len(open_array) > 0:
            current_node = None

            # it will find a node with the lowest value of f(n) = g(n) + h(n)
            for node in open_array:
                if current_node == None or node_distance[node] + self.heurestic_function(node) < node_distance[current_node] + self.heurestic_function(current_node):current_node = node

            # Corner condition
            if current_node == None:
                print('PATH DOES NOT EXIST!')
                return None

            # if node is goal state node
            if current_node == end:
                reconstruct_path = []

                # backtrack and get the exact path
                while ajacent_node_name[current_node] != current_node:
                    reconstruct_path.append(current_node)
                    current_node = ajacent_node_name[current_node]

                reconstruct_path.append(start)

                reconstruct_path.reverse()

                print(*reconstruct_path, sep="->")
                print()
                return reconstruct_path

            # Implementing priority queue and storing it in open array and closed array
            for (m, weight) in self.find_all_neighbours(current_node):
                if m not in open_array and m not in close_array:
                    open_array.add(m)
                    ajacent_node_name[m] = current_node
                    node_distance[m] = node_distance[current_node] + weight

                else:
                    if node_distance[m] > node_distance[current_node] + weight:
                        node_distance[m] = node_distance[current_node] + weight
                        ajacent_node_name[m] = current_node

                        if m in close_array:
                            close_array.remove(m)
                            open_array.add(m)

            open_array.remove(current_node)
            close_array.add(current_node)

        print('Path does not exist!')
        return None

    # function to return shortest path cost
    def path_cost(self, final_path):
        cost = 0
        for i in range(0, len(final_path) - 1):
            for j in adjacency_matrix[final_path[i]]:
                if final_path[i + 1] == j[0]:
                    cost += j[1] + self.heurestic_function(j[0])
        print("SHORTEST PATH COST : {}".format(cost))
        return cost


if __name__ == "__main__":
    adjacency_matrix = {
        'A': [('Z', 75), ('S', 140), ('T', 118)],
        'Z': [('O', 71)],
        'S': [('F', 99),('R',80)],
        'T': [('L', 111)],
        'L': [('M', 70)],
        'O': [('S', 151)],
        'F': [('B', 211)],
        'R': [('P', 97)],
        'M': [('D', 75)],
        'P': [('B', 101)],
        'D': [('C', 120)],
        'C': [('R', 146),('P',138)],
        'T': [('L', 111)],
    }
    graph1 = Graph(adjacency_matrix)
    print("\nA* SEARCH ALGORITHM\t\t\n")
    print("SHORTEST PATH FROM S(source) - D(destination) : - \t\n")
    shortest_path = graph1.a_star_main('A', 'B')
    shortest_path_cost = graph1.path_cost(shortest_path)
```
