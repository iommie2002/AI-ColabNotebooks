```python
import numpy as np
import pandas as pd

df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.tail()
```

```python
column_data = df["Age"]
print(column_data)
```

```python


age_column = df["Age"]


min_range = 0
max_range = 1


min_age = age_column.min()
max_age = age_column.max()
normalized_age = (age_column - min_age) / (max_age - min_age) * (max_range - min_range) + min_range


normalized_df = df.copy()
normalized_df["Normalized_Age"] = normalized_age


print(normalized_df.head())
```

```python
age_column = df["Age"]

# Calculate the mean, median, mode, and standard deviation of the "Age" column
mean_age = np.mean(age_column)
median_age = np.median(age_column)
mode_age = age_column.mode()[0]
std_deviation_age = np.std(age_column)

# Standardize the "Age" column
z_score_age = (age_column - mean_age) / std_deviation_age

# Create a DataFrame to store the results
normalized_df = pd.DataFrame({
    "Original_Age": age_column,
    "Mean_Normalized_Age": (age_column - mean_age) / std_deviation_age,
    "Median_Normalized_Age": (age_column - median_age) / std_deviation_age,
    "Mode_Normalized_Age": (age_column - mode_age) / std_deviation_age
})

# Display the normalized DataFrame
print(normalized_df.head())



```
