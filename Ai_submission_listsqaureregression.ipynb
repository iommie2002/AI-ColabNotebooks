```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax

from google.colab import files

uploaded = files.upload()
# Load your data from a CSV file
data = pd.read_csv('data_for_lr.csv')

```

```python
# Training dataset and labels
train_input = np.array(data.x[0:500]).reshape(500, 1)
train_output = np.array(data.y[0:500]).reshape(500, 1)

# Calculate the means of input and output
mean_x = np.mean(train_input)
mean_y = np.mean(train_output)

# Calculate the parameters θ1 (slope) and θ0 (intercept)
numerator = np.sum((train_input - mean_x) * (train_output - mean_y))
denominator = np.sum((train_input - mean_x) ** 2)
theta1 = numerator / denominator
theta0 = mean_y - theta1 * mean_x

# Define a function for linear regression
def linear_regression(x):
    return theta0 + theta1 * x

# Calculate the predictions for the training data
predictions = linear_regression(train_input)

# Calculate Mean Squared Error (MSE)
n = len(train_output)
mse = (1/n) * np.sum((train_output - predictions) ** 2)

# Calculate Mean Absolute Error (MAE)
mae = (1/n) * np.sum(np.abs(train_output - predictions))

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)

# Print the equation of the line of linear regression
print("Linear Regression Line: y = {:.4f}x + {:.4f}".format(theta1, theta0))

# Plot the regression line
plt.scatter(train_input, train_output, label='Data')
plt.plot(train_input, predictions, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
